```python
train_X, train_y = train_test_split(df, train_size=0.7, random_state=111)
train_X.head()
```

```python
len(train_X)
```

```python
len(train_y)
```

## 5. model

```python
model.score(train_X[['Pick_rate', 'Tie_Rate', 'On_fire']], train_X[['Win_rate']])
```

```python
model.coef_
```

```python
model.intercept_
```

### --LogisticRegression--

```python
df2.head()
```

```python
df5 = df2[df2['Rank'].str.match('All|Grandmaster')]
df5
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
X, y = df5[['Win_rate']], df5.Rank
test_size=0.3
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=7)
```

```python
model = LogisticRegression()
model
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
model.coef_
```

```python
model.predict(X_test)
```

```python
model.predict_proba(X_test)
```

```python
predicted = pd.DataFrame(model.predict(X_test), columns={'predicted':''})
predicted
```

```python
y_test = pd.DataFrame(y_test)
y_test.shape
```

```python
y_test['predicted'] = predicted
```

```python
y_test = y_test.reset_index()
y_test
```

```python
pd.concat([y_test, predicted], axis=1)
```

```python
pd.crosstab(y_test['Rank'], predicted['predicted'])
```

```python
from sklearn import metrics
```

```python
metrics.confusion_matrix(y_test['Rank'], predicted['predicted'])
```

```python
tn, fp, fn, tp = metrics.confusion_matrix(y_test['Rank'], predicted['predicted']).ravel()
```

```python
print(tn)
print(fp)
print(fn)
print(tp)
```

$$Accuracy={\frac {{TP+TN} }{TP + TN +FP + FN} }\\
Precision={\frac {TP}{TP+FP}}\\
Recall={\frac {TP}{TP+FN}}\\
F1 = 2 \times {\frac {precision \times recall} {precision + recall}}
$$
